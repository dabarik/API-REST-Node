```python
#les imports
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
temperature = ctrl.Antecedent(np.arange(0,40,1), "temperature")
temp_clim = ctrl.Antecedent(np.arange(-1,1,1), "temp_clim")
```

```python
temp_tres_froid = fuzz.sigmf(temperature.universe, 5, -0.8)

temp_froid = fuzz.gbellmf(temperature.universe, 5, 1.5, 7)

temp_chaleureux = fuzz.gbellmf(temperature.universe, 7, 1.5, 15)

temp_chaud = fuzz.gbellmf(temperature.universe, 20, 1.5, 25)

temp_tres_chaud = fuzz.sigmf(temperature.universe, 25, 0.8)

temperature['tres_froid'] = temp_tres_froid
temperature['froid'] = temp_froid
temperature['chaleureux'] = temp_chaleureux
temperature['chaud'] = temp_chaud
temperature['tres_chaud'] = temp_tres_chaud

plt.plot(temperature.universe, temp_tres_froid,  label='tres_froid')
plt.plot(temperature.universe, temp_froid,  label='froid')
plt.plot(temperature.universe, temp_chaleureux,  label='chaleureux')
plt.plot(temperature.universe, temp_chaud,  label='chaud')
plt.plot(temperature.universe, temp_tres_chaud,  label='tres_chaud')

plt.legend()
plt.show()
```
